# # # Calculating the percentage of confirmed cases regarding the Saudi Arabia population

```python
df_KSA['population'].unique()
```

```python
df_KSA['total_cases'].max()
```

VISUALLIZING WORLD DATA
Confirmed cases (Total Cases) Saudi Arabia

```python
df_KSA_over_time = df_KSA.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index().sort_values("date",ascending=True).reset_index(drop=True)
df_KSA_over_time
```

```python
# Using line-plot
plt.figure(figsize=(10,5))
plt.plot(df_KSA_over_time.index, df_KSA_over_time['total_cases'])
plt.title('Evolution of Confirmed Covid-19 cases over time in Saudi Arabia', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Confirmed cases', fontsize=16)
```

# Total deaths cases evolution over time

```python
# Using line-plot
plt.figure(figsize=(15,5))
plt.plot(df_KSA_over_time.index, df_KSA_over_time['total_deaths'])
plt.title('Evolution of Covid-19 Deaths cases over time in Saudi Arabia', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Number of Deaths', fontsize=16)
```

# New cases  Saudi Arabia

```python
# Using line-plot
plt.figure(figsize=(15,7))
plt.bar(df_KSA_over_time.index, df_KSA_over_time['new_cases'])
plt.title('Evolution of Covid-19 New Cases over time in Saudi Arabia', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
```

# ****Putting it all together

```python
# Using line-plot
plt.figure(figsize=(15,5))
plt.plot(df_KSA_over_time.index, df_KSA_over_time['total_cases'], label='Confirmed')
plt.plot(df_KSA_over_time.index, df_KSA_over_time['total_deaths'], label='Total Deaths')
plt.plot(df_KSA_over_time.index, df_KSA_over_time['new_cases'], label='New')
plt.plot(df_KSA_over_time.index, df_KSA_over_time['new_deaths'], label='New Deaths')
plt.legend(loc=0)
plt.show()
```
